```python
def clean_buyer_comments(x):
  if pd.isnull(x):
    return "No Comments"
  x = x.lower()

  if any ( word in x for word in ['loved','excellent','great',]):
    return " Positive"
  elif any( words in x for words in ['not','far','underpriced','poor']):
    return "Negative"
  else:
    return "Neutral"

data['Buyer_sentiment']=data['Buyer_Comments'].apply(clean_buyer_comments)
```
