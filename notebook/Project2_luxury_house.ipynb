Market Trends


How have luxury housing bookings changed quarter by quarter across micro-markets?
 → (Line Chart with Quarter on X-axis and Booking_Count on Y-axis, segmented by Micro_Market)


```python
cursor.execute("""
SELECT Booking_Status, COUNT(*)
FROM housing_db
GROUP BY Booking_Status
""")
print(cursor.fetchall())
```

```python
cursor.execute("""DROP TABLE IF EXISTS market_trends;""")
cursor.execute("""
CREATE TABLE market_trends AS
SELECT
    Year,
    Quarter,
    Micro_Market,
    COUNT(*) AS Booking_Count
FROM housing_db
WHERE Luxury_Price_Flag = 1
GROUP BY Year, Quarter, Micro_Market
ORDER BY Year, Quarter;
""")

conn.commit()
```

```python
cursor.execute("SELECT * FROM market_trends LIMIT 10")
print(cursor.fetchall())
```

```python
df_market = pd.read_sql("SELECT * FROM market_trends", conn)
df_market.to_csv("market_trends.csv", index=False)
```

Booking Conversion


Which micro-markets have the highest and lowest booking conversion rates?
 → (Stacked Column Chart: Micro_Market with % Booking_Status as stacked sections)


```python
data.info()
data.head()
```

```python
cursor.execute("""DROP TABLE IF EXISTS Conversion_rates;""")
cursor.execute("""
CREATE TABLE Conversion_rates AS
SELECT
    Micro_Market,

    COUNT(*) AS Total,

    SUM(CASE
        WHEN Booking_Status = 'Booked' THEN 1
        ELSE 0
    END) AS Booked,
    SUM(CASE
        WHEN Booking_Status = 'Not Booked' THEN 1
        ELSE 0
    END) AS Not_Booked,

    CAST(SUM(CASE
        WHEN Booking_Status = 'Booked' THEN 1
        ELSE 0
    END) AS FLOAT) / COUNT(*) AS Conversion_Rate

FROM housing_db

GROUP BY Micro_Market

ORDER BY Conversion_Rate DESC;
""")
```

```python
cursor.execute("SELECT * FROM Conversion_rates")
rows = cursor.fetchall()

for row in rows:
    print(row)
```

Possession Status Analysis

```python
data[['Possession_Status','Buyer_Type','NRI_Buyer']].head()
```

```python
cursor.execute("""DROP TABLE IF EXISTS possession_analysis;""")
cursor.execute("""
CREATE TABLE possession_analysis AS
SELECT
    Possession_Status,
    Buyer_Type,

    COUNT(*) AS Total_Leads,

    SUM(CASE
        WHEN Booking_Status = 'Booked' THEN 1
        ELSE 0
    END) AS Booked_Count,

    SUM(CASE
        WHEN Booking_Status = 'Not Booked' THEN 1
        ELSE 0
    END) AS Not_Booked_Count,

    CAST(SUM(CASE
        WHEN Booking_Status = 'Booked' THEN 1
        ELSE 0
    END) AS FLOAT) / COUNT(*) AS Conversion_Rate

FROM housing_db

GROUP BY Possession_Status, Buyer_Type

ORDER BY Possession_Status, Conversion_Rate DESC;
""")

conn.commit()
```

Top Performers


Who are the top 5 builders in terms of revenue and booking success?
 → (Card visuals or KPI indicators with drill-through to detailed project list

```python
cursor.execute("""DROP TABLE IF EXISTS Top_Performers;""")

cursor.execute("""CREATE TABLE Top_Performers AS
SELECT  Developer_Name , SUM(Ticket_Price_Cr) AS Total_Revenue,
SUM(CASE
        WHEN Booking_Status = 'Booked' THEN 1
        ELSE 0
    END) AS Booked_Count,

    CAST(SUM(CASE
        WHEN Booking_Status = 'Booked' THEN 1
        ELSE 0
    END) AS FLOAT) / COUNT(*) AS Conversion_Rate
    FROM housing_db

GROUP BY Developer_Name

ORDER BY Total_Revenue DESC
 """)
conn.commit()
```
